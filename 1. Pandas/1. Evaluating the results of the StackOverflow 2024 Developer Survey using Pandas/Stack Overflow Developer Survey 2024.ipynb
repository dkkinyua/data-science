```python
# Let's get the Kenyan devs who know Python
df.loc[filt]["LanguageHaveWorkedWith"].str.contains("Python").sum()
 
# 106 Kenyan devs know Python in this survey!
```

```python
# Let's get this info using country_grp
country_grp["LanguageHaveWorkedWith"].apply(lambda x: x.str.contains("Python").sum()).loc[["Kenya", "South Africa"]]
```

```python
# We want to get the percentages of devs who use Python that answered this survey. To do so, we need to add the Python devs and all devs into one dataframe
# Get all devs that use Python
python_devs = country_grp["LanguageHaveWorkedWith"].apply(lambda x: x.str.contains("Python").sum())
python_devs
```

```python
# Let's get all devs who answered this survey according to Country
all_devs = df["Country"].value_counts()
all_devs.head(50)
```

```python
# Let's join this together into a dataframe using the Pandas.concat() method
python_df = pd.concat([all_devs, python_devs], sort=False, axis="columns") # We can use axis=1 to align data along the columns.
python_df
```

```python
# Let's rename the column names
python_df.rename(columns={"count":"All Devs", "LanguageHaveWorkedWith":"All Devs that use Python"}, inplace=True)
python_df.head(70)
```

```python
# Now, let's get the pct who know Python that answered this survey
python_df["Pct of Devs who Know Python"] = (python_df["All Devs that use Python"] / python_df["All Devs"]) * 100
python_df
```

```python
# Let's get the pct of Kenyan developers who use Python that answered this survey
python_df.loc["Kenya"]
```
