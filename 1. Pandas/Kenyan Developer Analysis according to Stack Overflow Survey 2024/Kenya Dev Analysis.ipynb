# Visualizing programming languages used by Kenyan developers.

I would like to get what languages Kenyan developers use in their daily work and visualize them into a bar chart.
These languages might be many, of course, but I'd like to narrow down to the most commonly used languages globally like Python, Javascript, Typescript, Java, Go, Kotlin among others

```python
df["LanguageHaveWorkedWith"]
```

```python
df["LanguageHaveWorkedWith"].unique()
```

```python
langs_list = ["Python", "Kotlin", "Go", "Ruby", "Rust", "C", "C++", "JavaScript", "TypeScript", "C#", "PHP", "Dart", "Scala", "R", "Lua", "HTML/CSS", "SQL", "Java", "Swift"]
votes = []

for i in langs_list:
    filt = (df["LanguageHaveWorkedWith"].str.contains(i, na=False)).value_counts()
    count = filt.get(True, 0)
    votes.append(count)
votes
```

```python
import matplotlib.pyplot as plt

plt.title("A graph showing different programming languages used by Kenyan developers in 2024")
plt.xlabel("Number of developers")
plt.xticks(rotation=90) # Rotates the labels on the x-axis by 45deg to avoid overlapping
plt.ylabel("Languages used")
plt.bar(langs_list, votes)
```

# Visualizing the working arrangements used by Kenyan developers who took this survey

Let's visualize the working arrangements used by Kenyan developers who took this survey, that is, if they work fully remote, on site or have a hybrid working environment using a pie chart.

```python
df.head(10)
```

```python
# Let's create a list of the values of the arrangements
work = list(df["RemoteWork"].value_counts())
work
```

```python
arrangements = ["Hybrid", "Remote", "In-person"]

plt.title("A chart showing the working arrangement of some Kenyan developers")
plt.pie(work, labels=arrangements, autopct='%.1f%%')
```
