```python
# This gets all rows/participants who have a higher salary than $30k
highest_salary = df["ConvertedCompYearly"] > 30000
df.loc[highest_salary]
```

```python
# Let's get all participants who work with Python and earn $30k or more
filt = (df["LanguageHaveWorkedWith"] == "Python") & (df["ConvertedCompYearly"] > 30000)
df.loc[filt, ['LanguageHaveWorkedWith', 'ConvertedCompYearly', 'Country']]
```

```python
# This function helps us in getting the rows with the folwing countries stated below
countries = ["United States of America", "Germany", "France", "Italy"]
filt2 = df["Country"].isin(countries)
df.loc[filt2]
```

```python
# gets the rows with languages == Python. NB: Don't forget to add na=False to filter out the NaN values 
filt3 = df["LanguageHaveWorkedWith"].str.contains("Python", na=False)
df.loc[filt3]
```
